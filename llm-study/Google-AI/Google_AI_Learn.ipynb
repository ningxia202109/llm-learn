Get Key from secret

```python
import google.generativeai as genai
from google.colab import userdata
GOOGLE_AI_KEY=userdata.get('GOOGLE_AI_KEY')
genai.configure(api_key=GOOGLE_AI_KEY)
```

Run prompte



```python
model = genai.GenerativeModel('gemini-pro')
response = model.generate_content("Give me python code to sort a list")
print(response.text)
```

Load image from content /content/images/vancouver-pic.jpg

```python
import PIL.Image
img = PIL.Image.open('/content/images/vancouver-pic.jpg')
img
```

Ask question for above image

```python
prompt = """
This image has lots of items, please describe the summary of this image and then list all items that you can see, return output in json format:
{description: description, items: [items1, items2, items3, etc]}
"""

model = genai.GenerativeModel('gemini-pro-vision')
response = model.generate_content([prompt, img])
print(response.text)
```

CHat with gemini

```python
model = genai.GenerativeModel('gemini-pro')
chat = model.start_chat(history=[])
response = chat.send_message("In one sentence, explain how a computer works to a young child.")
print(response.text)
```
